```python
import yfinance as yf
import pandas as pd
import seaborn as sns
```

```python
ações = ['BPAC11.SA', 'PETR4.SA', '^BVSP', 'OIBR3.SA']

preços = yf.download(ações)
preços.dropna(inplace=True)
```

```python
preços_fechamento = preços.Close
```

```python
matriz_corr = preços_fechamento.corr()
sns.heatmap(matriz_corr)
```
